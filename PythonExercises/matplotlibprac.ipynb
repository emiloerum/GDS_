```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.1)
y = np.cos(x)
plt.plot(x, y, label='plot.png')
#always call savefig before show or else it wont save the figure
plt.savefig('plot.png', transparent=True)
#legend reads the labels of the plots and shows them
#in the top right corner. Good for multiple plots in the same window
plt.legend()
plt.show()
#Do savefig to save the plot
```

```python
#samples of a standard normal distribution
x = np.random.standard_normal(10000)
#the optional argument bins below is the amount
#of cols in the histogram
plt.hist(x, bins=100)

plt.show()
```

```python
#2D gaussian distribution sample
x = np.random.multivariate_normal(
        mean=(1,2),
        cov=[[1, -0.5], [-0.5, 1]],
        size=1000)

#alpha is the opacity (how many things are overlapping)
plt.scatter(x[:,0],x[:,1], alpha=0.5)
plt.show()
```

```python
#SUBPLOTS/ Multiple plots in the same figure
#Linearly spaced x values
#num argument is the number of samples to generate
x = np.linspace(0.0, 5.0, num=100)

#creating y values
y1 = np.cos(2* np.pi* x)
y2 = np.cos(2* np.pi* x) * np.exp(-x)

#Plotting
plt.subplot(2, 1, 1) #2 rows, 1 column, index 1 (the top plot)
plt.plot(x, y1, '.-') # the '.-' argument specifies the format
#here that the elements should be dots . with lines - connecting them

plt.subplot(2, 1, 2) #2 rows, 1 column, index 2 (the bottom plot)
plt.plot(x, y2, '.-')
```
