```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table('british-english', keep_default_na=False)
df = df.rename(columns={'A' : 'words'})
print(df)
```

```python
rows_starting_with_A = df.loc[df['words'].str.startswith('A')]
print(rows_starting_with_A)
```

```python
#Grouping data by first letter in the each word in the column words:
groups = df.groupby(df['words'].str[0])
print(groups)

#printing the name of the groups and the groups:
for name, grp in groups:
    print(name)
    print(grp)

print("Amount of groups: ", len(groups))

#printing amount of words in each group
#to show counts for all groups click on the truncate thingy :P
words_in_groups = groups.agg(words_count =('words', np.size))
print(words_in_groups)

words_in_groups.plot.bar(legend='Words Count', xlabel='First Letter of Word', ylabel ='count of words',)
```
